### Создание попарных ксоров

Попарный ксор шифротекстов равен попарному ксору исходных сообщений. Эта информация дает представление о зависимостях между разными текстами: если, к примеру, представить, что в определенном месте ксора шифротекстов стоит бит 1, а по предположению в одном из текстов на этом месте стоит 0, то у второго текста обязана стоять 1 в том же месте.

Такая информация полезна для сужения множества допустимых значений символов в определенные места оригинальных текстов. Действительно, полагаясь на некоторую метаинформацию (к примеру, что преимущественное большинство допустимых символов оригинальных текстов - это цифры, заглавные и строчные английские буквы и символы пунктуации), можно некоторым образом перебирать эти символы в определенном тексте и отметать те варианты, где в остальных местах будут символы вне этого множества. Так можно определить некоторые символы в определенных позициях трех текстов однозначно и одновременно, а в остальных позициях останется или некоторая неопределенность из некоторых наборов троек символов, или отсутствуют предполагаемые тройки в целом.

Множество допустимых символов оригинальных текстов можно менять: при увеличении мы сможем увеличить вероятность получить хотя бы одну тройку символов на определенные позиции, однако рискуем получить позиции с некоторыми вариантами троек. При уменьшении размера мы можем оказаться в ситуации, когда будет мало позиций с одним набором тройек для определенных позиций трех исходных текстов. Кроме того, надо учитывать тот вариант, что брать набор рассматриваемых символов слишком малым тоже рискованно, поскольку мы не даем  гарантий, что при отказе от некоторых символов мы откажемся от тех, что действительно есть в тексте, однако на тех позициях могут быть 

```python
xors = []

for i in range(len(datas) - 1):
    each_xor = []
    for j in range (i+1, len(datas)):
        xor_res = bytes(a ^ b for (a, b) in zip(datas[i], datas[j]))
        print(xor_res)
        each_xor.append(xor_res)
    xors.append(each_xor)
```

### Поиск допустимых троек

Ниже представлен фрагмент кода (функция), которая по множеству ксоров определенных символов между одним текстом и остальными ищет допустимые значения.

По свойству ксора и условию, что все шифротексты получены ксором с одним и тем же ключом, то можно передать списиок ксоров символов на определенной позиции только из первого текста с остальными (ксор символов 2 и 3 текста между собой передавать, как и обрабатывать - бессмысленно). 

```python
def find_possible_symbols_sets(xors_res, possible_symbols):
    sets = []
    for symbol in possible_symbols:
        new_set = [symbol]
        is_possible = True
        
        for xor_res in xors_res:
            new_symbol = ord(symbol) ^ xor_res
            #print(symbol, ord(symbol), new_symbol)
            if chr(new_symbol) in possible_symbols:
                new_set.append(chr(new_symbol))
            else:
                is_possible = False
                break
        
        if is_possible:
            sets.append(new_set)
    return sets
```

Потому как первый шаг для перебора возможных значений логичным было бы использование информации о самом языке (в данном случае - английского): вероятность встречаемости ASCII символов, словарь английских слов, учет регистра символов и тд и тп. ### Подгрузка информации о распределении символов и словаря английских слов

```python
from nltk.corpus import words
vocabulary = set([str.lower(word) for word in words.words()])
print(len(vocabulary))
```

```python
import string      # definitions of ascii printable chars
from collections import defaultdict     # fast counting
import typing as tp

def generate_frequencies(text: str) -> tp.Dict:
    d = defaultdict(int)    # define dictionary for counting frequencies

    for ch in text:       # loop over each character 
        if ch in string.printable:     # is the character in the ascii/printable set?
            d[ch] += 1    #   if so, add 1 to that characters frequency counter
    return d

ref_files = [
    "./t8.shakespeare.txt",
    #"./war_and_peace.txt"
]

total_txt = ""
for path in ref_files:
    file = open(path, "r")
    data = file.read()
    print(f"{path}, content: {data[0:200]}")
    total_txt += data
    
tmp_d = generate_frequencies(total_txt)
print(tmp_d.keys(), len(tmp_d.keys()))     # print all frequencies

d = {}
counts = 0
for key in tmp_d.keys():
    if key in all_symbols:
        d.update({key: tmp_d[key]})
        counts += tmp_d[key]
print(d.keys(), len(d.keys()))     # print all frequencies
print(counts)

min_freq = 1
for key in d.keys():
    d[key] /= counts
    min_freq = min(min_freq, d[key])

for symbol in all_symbols:
    if symbol not in d.keys():
       d.update({symbol: min_freq / 2}) 
print(d)     # print all frequencies
```

```python
def map_triple(triple: tp.List[str]) -> int:
    nums = [ord(triple[0]), ord(triple[1]), ord(triple[2])]
    nums.sort()
    #print(nums)
    res = nums[0] * 256 ** 2 + nums[1] * 256 + nums[2]
    return res
    

# Составляем вероятности для троек символов на одну позицию в трех текстах сразу: поскольку текста независимы, то и символы в некоторых позициях можно считать независимыми
d_triple = {}
for i in range(0, len(all_symbols)):
    for j in range(i, len(all_symbols)):
        for k in range(j, len(all_symbols)):
            d_triple.update({map_triple([all_symbols[i], all_symbols[j], all_symbols[k]]): d[all_symbols[i]] * d[all_symbols[j]] * d[all_symbols[k]]})
print(len(d_triple.keys()))
```

Начало перебора с учетом представленной информации

Перебираем по такому принципу: начиная с некоторого символа подбираем на позицию наиболее вероятную тройку из кандидатов, потом пытаемся составить слово из английского словаря в одном из текстов путем перебора следующих символов. Проверяем что lower-case версии слов (или их частей) содержатся хотя бы в 1 слове из словаря. Если такое правило соблюдается, то будем считать, что кандидатная тройка на след символы определены. Если же при подборе тройки символов на все три текста удовлетворяющих троек не осталось, значит считаем, что предыдущий подбор был некорректен. А значит, нужно вернуться к предыдущему символу и попытаться построить все с начала.

```python
import re

def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

def is_text_valid(text: str) -> bool:
    
    
    # Проверка на то, что в след за идущими символвами пунктуации идут отступы
    for i in range(0, len(text) - 1):
        if text[i] in [',', '.', '!', '?', ':', ';', '\''] and text[i + 1] not in (' ', '\t', '\n'):
            return False
    
    # Проверка на то, что подряд идущие символы английского образуют читабельные слова
    
    # делим строку на слова согласно разделителям
    regex_pattern = '|'.join(map(re.escape, delimiters))
    text_splits = re.split(regex_pattern, text, maxsplit=0)
    
    # удаляем пустые строки
    text_splits = [word for word in text_splits if word != ' ' and word != '']
    
    #print(f"---CHECK NEXT TEXT: {text}. SPLITS ON '{text_splits}'")
    
    # проверяем, что заглавные символы слов не в середине слов
    for text_split in text_splits:
        word_to_check = text_split[1:]
        if len(word_to_check) == 0:
            continue
        
        isupper = word_to_check.isupper()
        islower = word_to_check.islower()
        if isupper == False and islower == False:    # слово без первой буквы должно быть или БОЛЬШИИИИМ, или млньким
            return False
        elif isupper == True and text_split[0].islower() == True:
            return False
        elif islower == True and text_split[0].isupper() == True:   # по случайности было обнаружено, что при переборе отбросилась фамилия Claypo... (подозреваю, что это имя собственное. Будем полагать что оно может быть любым)
            continue

    # проверяем, что цифр нет в середине строк
    for text_split in text_splits:
        if str.isdigit(text_split) == False and has_numbers(text_split): # Если слово целиком не число и при этом содержит цифры в себе - ошибка
            return False
    
    # проверяем, что образованные сочетания букв - части существующих слов английского
    for i, text_split in enumerate(text_splits):
        has_word = False
        for s in vocabulary:
            # если первое или последнее слово - проверяем на возможность быть частью других слов
            if i == 0 or i == len(text_splits) - 1:
                if str.lower(text_split) in str(s):
                    has_word = True
                    break
                elif str.isupper(text_split[0]) == True and str.islower(text_split[1:]) == True:
                    has_word = True
                    break
            # другие слова - должны быть из словаря или именами собственными
            else:
                if str.lower(text_split) == str(s):
                    has_word = True
                    break
                elif str.isupper(text_split[0]) == True and str.islower(text_split[1:]) == True:
                    has_word = True
                    break
        if has_word == False:
            return False

    return True
```

```python
# НЕКОТОРЫЕ ЗНАЧЕНИЕ МЕНЯЮТСЯ РУКАМИ ПОСКОЛЬКУ ПО ИТОГУ ВЫВОДА ОПРЕДЕЛЯЛИСЬ НЕКОТОРЫЕ ЗАКОНОМЕРНОСТИ ТЕКСТА, КОТОРЫЕ ПОТОМ НАРУШАЛИСЬ ПРИ НЕПРАВИЛЬНОЙ ПРОВЕРКЕ ТЕКСТА
# ДАННОЕ РЕШЕНИЕ ПРЕДПОЛАГАЕТ НЕКОТОРУЮ МОДЕДЕРАЦИЮ

START_POS = 0
LENGHT = 200

answer_stack = []
res1, res2, res3 = "", "", ""
start_set = possible_sets[START_POS]
start_set.sort(key=lambda x: d_triple[map_triple(x)])
for possible_set in start_set:
    answer_stack.append([START_POS, possible_set, res1, res2, res3])

print(answer_stack)

while len(answer_stack) >= 0:
    possible_state = answer_stack.pop()
    
    possible_text_1 = possible_state[2] + possible_state[1][0]
    possible_text_2 = possible_state[3] + possible_state[1][1]
    possible_text_3 = possible_state[4] + possible_state[1][2]
    
    is_1st_valid = is_text_valid(possible_text_1)
    is_2nd_valid = is_text_valid(possible_text_2)
    is_3rd_valid = is_text_valid(possible_text_3)
    
    str_output = ""
    if is_1st_valid == True:
        str_output += f"VALID : '{possible_text_1}' "
    else:
        str_output += f"FAILED: '{possible_text_1}' "
        
    if is_2nd_valid == True:
        str_output += f"VALID : '{possible_text_2}' "
    else:
        str_output += f"FAILED: '{possible_text_2}' "
    
    if is_3rd_valid == True:
        str_output += f"VALID : '{possible_text_3}' "
    else:
        str_output += f"FAILED: '{possible_text_3}' "
    print(str_output)
    
    if is_1st_valid == True and is_2nd_valid == True and is_3rd_valid == True:
        next_pos = possible_state[0] + 1
        if (next_pos - START_POS > LENGHT):
            print("---POSSIBLE ANSWERS---")
            print(f"text1:{possible_text_1}")
            print(f"text2:{possible_text_2}")
            print(f"text3:{possible_text_3}")
            continue
        
        next_pos_sets = possible_sets[next_pos]
        next_pos_sets.sort(key=lambda x: d_triple[map_triple(x)])
        for next_pos_set in next_pos_sets:
            answer_stack.append([next_pos, next_pos_set, possible_text_1, possible_text_2, possible_text_3])
    else:
        continue
```
